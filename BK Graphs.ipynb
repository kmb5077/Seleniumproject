```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
import math

rawdf = pd.read_csv("aptsBK.csv",header=None)
rawdf.columns = ['Address', 'Neighborhood','Price','Apt Type','Bedrooms','Bathrooms','Rooms','Sqft','Description']
```

```python
rawdf
```

```python
df1=rawdf.dropna(axis=0)
```

```python
df1.shape
```

```python
df1["Price"]=list(map(lambda x: x.replace(',',''.strip()),df1["Price"]))
df1["Price"]=list(map(lambda x: x.replace('$',''.strip()),df1["Price"]))
df1["Price"]=df1["Price"].astype('int')
df1["Sqft"]=list(map(lambda x: x.replace(',',''.strip()),df1["Sqft"]))
df1["Sqft"]=df1["Sqft"].astype('int')
df1["$/Sqft"]=df1["Price"]/df1['Sqft']
```

```python
df1
```

```python
plt.scatter(np.log(df1['Price']),df1['Neighborhood'])
plt.gcf().set_size_inches((15, 8))    
plt.show()
```

```python
df1_score = df1[['Neighborhood', 'Price']]
df1_score.boxplot(by='Neighborhood', column='Price')
plt.ylabel('Price')
```

```python
sns.boxplot(x='Neighborhood', y='Price', data=df1)
```

```python
df2=rawdf.dropna(axis=0)
df2["Price"]=list(map(lambda x: x.replace(',',''.strip()),df2["Price"]))
df2["Price"]=list(map(lambda x: x.replace('$',''.strip()),df2["Price"]))
df2["Price"]=df2["Price"].astype('int')
df2["Sqft"]=list(map(lambda x: x.replace(',',''.strip()),df2["Sqft"]))
df2["Sqft"]=df2["Sqft"].astype('int')
df2["$/Sqft"]=df2["Price"]/df2['Sqft']
df2['Log_Price']=df2['Price'].apply(math.log)
```

```python
df2boxplot=sns.boxplot(x='Bedrooms', y='Log_Price', data=df2,width=.8,)
plt.gcf().set_size_inches((15, 10))    
plt.show()
```

```python
df2boxplot=sns.boxplot(x='Log_Price', y='Neighborhood', data=df2,)
plt.gcf().set_size_inches((15, 10))    
plt.show()
```
